```python
def sum_range(start,end):
    if start > end:
        end, start = start, end
    return sum(range(start, end + 1))

print(sum_range(2,12))
print(sum_range(-4, 4))
print(sum_range(3,2))
```

```python
def fact(num):
    if num == 0:
        return 1
    else:
        return num * fact(num - 1)

print(fact(5))
```

```python
import numpy as np
def euclidan_distance(v1, v2):
    return sum((x-y) ** 2 for x, y in zip(v1,v2)) ** 0.5

x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
print(euclidan_distance(x,y))
```

```python
def sqr_euclidan_distance(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2))

def weighted_euclidan_distance(v1, v2, w):
    return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, w)) ** 0.5

def manhattan_distance(v1, v2):
    return sum(abs(x - y) for (x, y) in zip(v1, v2))

def chebyshev_distance(v1, v2):
    return max(abs(x - y) for (x, y) in zip(v1, v2))

x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
w = np.array([0, 0, 1])
print(sqr_euclidan_distance(x, y))
print(weighted_euclidan_distance(x, y, w))
print(manhattan_distance(x, y))
print(chebyshev_distance(x, y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = np.array([[-1, 1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

K = 3
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

print(' (-2,2) is class'),
print(model.predict([[-2, -2]]))

print(' (1,3) is class'),
print(model.predict([[1, 3]]))
```

```python
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

iris = sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(16,7))

plt.subplot(121)
sns.scatterplot(
    data = iris,
    x = 'petal_width', y = 'petal_length',
    hue = 'species',
    s = 70
)
plt.xlabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend()
plt.grid()

plt.subplot(122)
sns.scatterplot(data = iris, x = 'sepal_width', y = 'sepal_length', hue = 'species', s = 70)
plt.xlabel('–î–ª–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞, —Å–º')
plt.ylabel('–®–∏—Ä–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞, —Å–º')
plt.legend()
plt.grid();
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.20
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
X_train.head()
y_train.head()

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred

plt.figure(figsize=(10, 7))
sns.scatterplot(x = 'petal_width', y = 'petal_length', data = iris, hue = 'species', s = 70)
plt.xlabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend(loc = 2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color = 'red', s = 150)

print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')
```
